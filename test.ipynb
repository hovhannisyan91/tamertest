```python
    # Install Libraries
import pandas as pd
import numpy as np 
import spacy 
from spacy import displacy
import matplotlib.pyplot as plt
import scipy 
import re
```

```python

NER = spacy.load('en_core_web_sm')
```

```python

# Import Files
with open('20th_century.txt', 'r', errors='ignore') as file:
    data = file.read().replace('\n', '') 
```

```python
countries_df = pd.read_csv('countries.csv', index_col=False)
```

```python

country = NER(data)
```

```python
displacy.render(country[200:400], style='ent', jupyter=True)
```

```python
# Create list with sentences
sentence = [] # Create empty list

# Loop sentences
for sent in country.sents:
    entity_list = [ent.text for ent in sent.ents]
    sentence.append({'sentence':sent, 'entity':entity_list})
```

```python
# Create dataframe
sentence = pd.DataFrame(sentence)
sentence.head(10)
```

```python
# Filter using country list
def filter_entity(ent_list, countries):
    return[ent for ent in ent_list
          if ent in list(countries['country_name'])]
```

```python
sentence['c_entities'] = sentence['entity'].apply(lambda x: filter_entity(x, countries_df))
```

```python
sentence_filter = sentence[sentence['c_entities'].map(len)>0]
sentence_filter.head(10)
```
